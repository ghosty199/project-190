```python
#import the libraries


#Task 1
#Read the bmi.csv
import pandas as pd
from matplotlib import pyplot as plt

dataframe = pd.read_csv('bmi.csv')

dataframe
```

```python
#Task 2
#Data is sorted in descending order in accordance with BMI value
#Find the top 5 age group where the BMI value is the highest, and plot a bar graph out of it
df = pd.DataFrame(dataframe, columns =['Age','BMI'])
df.replace(" ", float("NaN"), inplace=True)

df = df.dropna()

sorted_df = df.sort_values(by=['BMI'],ascending=False)
sorted_df

heavy_satellite_5 = sorted_df.head(5)
print(heavy_satellite_5)

name = heavy_satellite_5['Age']
weight = heavy_satellite_5['BMI']

plt.xlabel("BMI")
plt.xticks(rotation='vertical')
plt.ylabel("Age")


label = name
value = weight 
plt.bar(label, value,width=0.4, color=('red','yellow','green','pink','blue')) #bar-grap
```
